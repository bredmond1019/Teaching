```python
# Note that we use a different item here. This will let us manipulate the 3D models
%matplotlib notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
startups = pd.read_csv("50_Startups.csv")
```

```python
# We need to make the data into arrays using numpy
# This is slightly different than the past

X = np.array(startups['R&D Spend'])
y = np.array(startups['Marketing Spend'])
z = np.array(startups['Profit'])
```

```python
fig = plt.figure(figsize = (8,4))
ax = fig.add_subplot(111, projection = '3d')
plt.title('R & D vs Marketing vs Profit')
ax.set_xlabel('R & D')
ax.set_ylabel('Marketing')
ax.set_zlabel("Profit")
ax.scatter(X, y, z)
```

```python
fig = plt.figure(figsize = (8,4))
ax = fig.add_subplot(111, projection = '3d')
plt.title('R & D vs Marketing vs Profit')
ax.set_xlabel('R & D')
ax.set_ylabel('Marketing')
ax.set_zlabel("Profit")
ax.plot(X, y, z)
```
